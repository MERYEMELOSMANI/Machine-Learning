```python
# Import libraries
import pandas as pd #importer la biblio nécessaire pour l'imporataion du fichier  csv 
import numpy as np #importer la biblio nécessaire pour les traitements mathématiques 
import matplotlib.pyplot as plt # importer la biblio nécessaire pour la visualisation des données 
from sklearn.model_selection import train_test_split #importer la biblio nécessaire pour diviser les données en ensembles d'entraînement et de test.
from pandas.core.common import random_state 
from sklearn.linear_model import LinearRegression #importer la biblio pour construire le modèle.
```

```python
# Get dataset
df_sal = pd.read_csv('Salary_Data.csv') #importer le fichier csv 
df_sal.head() #afficher les 5 lignes premiers 
```

```python
# Describe data
df_sal.describe() #description des donnnées 
```

```python
# Relationship between Salary and Experience
plt.scatter(df_sal['YearsExperience'], df_sal['Salary'], color = 'lightcoral') #visauliser le nuage de point a partir des donnees qui esxistent dans le fichier csv 
plt.title('Salary vs Experience') #le titre de visualisation
plt.xlabel('Years of Experience') #le labelle de la variable indépendante 
plt.ylabel('Salary') #le labelle de la variable dépendante 
plt.box(False) #n'affiche les axes des variables
plt.show() #visualise le nuage des points
```

```python
# Splitting variables
X = df_sal.iloc[:, :1]  # independent #indiquer la variable indépandante dans le fichier csv la deuxiéme colonne n'est pas inclues
y = df_sal.iloc[:, 1:]  # dependent #indiquer la variable dépandante dans le fichier csv la premiére  colonne n'est pas inclues
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) #diviser le dataset (20% de dataset va etre utilisé pour le test et random_state indique reproductibilité
```

```python
# Regressor model
regressor = LinearRegression() #appel du model 
regressor.fit(X_train, y_train) # fit fait l'ajustement du model sur les données d"entrainement
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)  #prédire la valeur de y_test   # predicted value of y_test
y_pred_train = regressor.predict(X_train)   #prédire la valeur de y_train  # predicted value of y_train
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'lightcoral') #visualiser le nuage de point des données d'entrainement 
plt.plot(X_train, y_pred_train, color = 'firebrick') #viualiser la droite de prédiction des données d'entrainement
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white') #la légende qui affiche a quoi sert la droite et les points
plt.box(False)
plt.show()
```

```python
# Prediction on test set
plt.scatter(X_test, y_test, color = 'lightcoral') #visualiser le nuage de point des données de test
plt.plot(X_train, y_pred_train, color = 'firebrick')  #viualiser la droite de prédiction des données de test
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}') #affiche le coefficient directeur 
print(f'Intercept: {regressor.intercept_}') #affiche l'intercept
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Parameters
num_records = 1000
start_date = datetime(2024, 1, 1, 0, 0, 0)
end_date = datetime(2024, 1, 7, 23, 59, 59)
location_ids = [1, 2, 3]
vehicle_types = ['car', 'truck', 'bus', 'motorcycle']
incident_types = ['none', 'minor accident', 'major accident', 'road work']
signal_phases = ['red', 'green', 'yellow']

# Generate random timestamps
timestamps = [start_date + timedelta(minutes=random.randint(0, int((end_date - start_date).total_seconds() / 60))) for _ in range(num_records)]

# Generate random data
data = {
    'timestamp': timestamps,
    'location_id': [random.choice(location_ids) for _ in range(num_records)],
    'traffic_volume': [random.randint(10, 100) for _ in range(num_records)],
    'average_speed': [random.uniform(20, 80) for _ in range(num_records)],
    'vehicle_type': [random.choice(vehicle_types) for _ in range(num_records)],
    'incident': [random.choice(incident_types) for _ in range(num_records)],
    'temperature': [random.uniform(-10, 35) for _ in range(num_records)],
    'precipitation': [random.uniform(0, 10) for _ in range(num_records)],
    'visibility': [random.uniform(0, 10) for _ in range(num_records)],
    'signal_phase': [random.choice(signal_phases) for _ in range(num_records)],
    'day_of_week': [ts.strftime('%A') for ts in timestamps],
    'time_of_day': ['morning' if 5 <= ts.hour < 12 else 'afternoon' if 12 <= ts.hour < 17 else 'evening' if 17 <= ts.hour < 21 else 'night' for ts in timestamps]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('synthetic_traffic_data.csv', index=False)

print('Synthetic traffic data generated and saved to synthetic_traffic_data.csv')
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Parameters
num_records = 10000  # Number of records to generate
start_date = datetime(2024, 1, 1, 0, 0, 0)
end_date = datetime(2024, 1, 7, 23, 59, 59)
location_ids = ['Times Square', 'Brooklyn Bridge', 'Central Park', 'Wall Street', 'Fifth Avenue']
vehicle_types = ['car', 'truck', 'bus', 'motorcycle']
incident_types = ['none', 'minor accident', 'major accident', 'road work']
signal_phases = ['red', 'green', 'yellow']

# Generate random timestamps
timestamps = [start_date + timedelta(minutes=random.randint(0, int((end_date - start_date).total_seconds() / 60))) for _ in range(num_records)]

# Generate random data
data = {
    'timestamp': timestamps,
    'location_id': [random.choice(location_ids) for _ in range(num_records)],
    'traffic_volume': [random.randint(10, 200) for _ in range(num_records)],
    'average_speed': [random.uniform(5, 60) for _ in range(num_records)],  # Speed in mph
    'vehicle_type': [random.choice(vehicle_types) for _ in range(num_records)],
    'incident': [random.choice(incident_types) for _ in range(num_records)],
    'temperature': [random.uniform(-10, 35) for _ in range(num_records)],  # Temperature in Celsius
    'precipitation': [random.uniform(0, 10) for _ in range(num_records)],  # Precipitation in mm
    'visibility': [random.uniform(0, 10) for _ in range(num_records)],  # Visibility in km
    'signal_phase': [random.choice(signal_phases) for _ in range(num_records)],
    'day_of_week': [ts.strftime('%A') for ts in timestamps],
    'time_of_day': ['morning' if 5 <= ts.hour < 12 else 'afternoon' if 12 <= ts.hour < 17 else 'evening' if 17 <= ts.hour < 21 else 'night' for ts in timestamps]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('nyc_synthetic_traffic_data.csv', index=False)

print('Synthetic NYC traffic data generated and saved to nyc_synthetic_traffic_data.csv')
```
